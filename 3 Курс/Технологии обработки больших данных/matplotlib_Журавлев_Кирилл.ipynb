1. Постройте график линейной функции (kx+b) с коэффициентами k и b, вводимыми пользователем. Предварительно рассчитать 100 значений аргумента из диапазона введенного пользователем.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
d1, d2 = map(int, input().split())
k, b = map(int, input().split())
y=[]
x=[]
for i in np.linspace(int(d1),int(d2), 100):
  x.append(i)
  y.append(int(k)*i+int(b))

plt.plot(x, y)
plt.show()
```

2. Постройте графики y = sin(x) и y = e-0,1xcos(x) на одной картинке. Сделайте их разными цветами и добавьте легенды.

```python
x = np.linspace(-10, 10, 100)
y1 = np.sin(x)
y2 = np.exp(-0.1 * x) * np.cos(x)
plt.plot(x, y1, label = "y = sin(x)")
plt.plot(x, y2, label = "y = exp(-0,1x)cos(x)")
plt.legend()
plt.show()
```

3. Постройте графики y = x2 и y = |x| на двух картинках в строке. Добавьте заголовки и подписи осей.

```python
y1 = [i*i for i in x]
y2 = [abs(i) for i in x]
fig, ax = plt.subplots(1,2, figsize = (20,10))
ax[0].plot(x,y1)
ax[0].set_xlabel("X")
ax[0].set_xlabel("Y1")
ax[0].set_title("y = x^2")
ax[1].plot(x,y2)
ax[1].set_xlabel("X")
ax[1].set_xlabel("Y2")
ax[1].set_title("y = |x|")
plt.show()
```

6. Сгенерируйте случайный процесс X(t) = r*X(t-1) + e(t), где дисперсия случайной нормальной добавки De = (1-r^2)*Dx. Dx = 1, r = 0.93. Отобразите его на графике. Сделайте на графике сетку.

```python
N = 1000
Dx = 1 
r = 0.9
De = (1-r**2)*Dx 
x = np.zeros(N)
x[0] = np.random.normal(0, np.sqrt(Dx))

for i in range(1, N):
  x[i] = x[i-1] * r + np.random.normal(0, np.sqrt(De))
df = pd.DataFrame({'x':x})

plt.plot(df) 
plt.title("X(t) = rX(t-1) + e(t)", fontsize=20)
plt.grid(color='black', alpha=0.8,linestyle = '--',linewidth = 1)
plt.show()
```

9. Построить секторную диаграмму для X из задания 6. Секторами диаграммы должно быть количество значений в интервале: [-inf, -2], [-2,0], [0, inf]. 

```python
fig, ax = plt.subplots()
vals = [0, 0, 0]
values = df['x']

for i in range(0, len(values)):
  if (values[i] <= -2):
    vals[0]+= 1
  elif (values[i] >= 0):
    vals[2] += 1
  else:
    vals[1] += 1

labels = ['[-inf, -2]',
          '[-2, 0]',
          '[0, +inf]']

ax.pie(vals, autopct = '%1.2f%%', labels = labels, colors = ['red', 'blue','green'])
ax.legend(loc='upper left', bbox_to_anchor=(1.0, 1.0),labels=labels)
plt.show()
```
